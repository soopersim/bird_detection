# Part A: Application Area Review

Object detection is one of the main areas where artificial intelligence has made great strides in imaging. Object detection is defined as the process of identifying the location of objects in a given image (object localisation) and the category to which each object belongs (object classification) (Zhao et al., 2019). Through the use of variety of domains, AI machines can including anomaly detection, autonomous driving, health care monitoring, and more. The performance of object detectors and trackers has significantly increased with the quick development of deep learning (DL) networks and GPU computing capability (Pal et al., 2021). With advancements in artificial intelligence (AI), numerous techniques have been developed to improve the accuracy and efficiency of object detection systems.


### Overview of AI Techniques in Object Detection
AI techniques have enabled systems to recognize complex patterns in visual data. The primary methods include:


1.   **Convolutional Neural Networks (CNNs)**: CNNs are widely used for image processing tasks. A CNN is made up of basic but non-linear modules that use its Convultional Layer, Pooling layer and fully connected layer to learn various levels of abstraction. Typically, arrays of pixel values from several colour channels, including Red, Green, and Blue (RGB), make up the input to the first layer. In general, the first layer determines whether edges with specific orientations are present or absent at particular points in the image. Later layers recognise sections of known objects by integrating motifs that are encoded into feature maps by detecting combinations of these edges. Ultimately, these components are put together in the final layers to detect entire objects (Moradi et al., 2024). Moreover, CNNs serve as the foundational architecture for many object detection models, including YOLO (You Only Look Once) and R-CNN (Region-based CNN).
2.   **Region-based Convolutional Neural Networks (R-CNNs)**: R-CNN models improve object detection by proposing regions of interest within an image and classifying these regions using CNNs. Improved models of R-CNN include, Fast R-CNN and Faster R-CNN. Fast R-CNN jointly optimizes classification and bounding box regression tasks. Faster R-CNN takes an additional subnetwork to generate region proposals (Zhao et al., 2019).
3. **You Only Look Once (YOLO)**: YOLO models treat object detection as a regression problem, predicting bounding boxes and class probabilities directly from an input image. This approach enables real-time performance by making all predictions in a single forward pass. YOLOv1 and YOLOv2 use a softmax function to convert class scores into probabilities, which works well for mutually exclusive classes. YOLOv3 introduces multi-label classification with an independent logistic classifier for each label. This eliminates the complexity of the softmax function and calculates loss using binary cross-entropy for each label (OpenAI, 2025).

  Key enhancements in YOLOv3 include:
  1. Bounding Box Optimization:
    
    Logistic regression predicts the presence of objects by calculating objectness scores for anchor boxes.
    
    Ground truth boxes are assigned to anchor boxes with the highest overlap, optimizing localization. Classification and localization losses are applied only to selected anchor boxes, while others incur only confidence loss. (- & Ravish, 2019)
    
  2. Feature Pyramid Network (FPN):
    
    YOLOv3 performs multi-scale predictions, generating bounding boxes, objectness scores, and class probabilities for each grid cell.

    This pyramid-like structure enables accurate detection of objects at varying scales, with predictions computed as, S*S*[3*(4+1+80)], where S is the grid size. (- & Ravish, 2019).

  This shows that by  leveraging CNNs as its backbone, YOLO demonstrates the power of AI in achieving real-time object detection with high efficiency and accuracy.





### Application in Object Detection
There are several uses for object detection that offer insightful information for semantic analysis of pictures and videos. It is extensively utilised in human behaviour analysis to analyse actions in diverse situations and in picture classification, where it recognises and classifies particular items within visual input. In order to discover faces prior to authentication or analysis, face recognition systems also rely on object detection as a fundamental step. Beyond these uses, improvements in learning systems and neural network algorithms keep improving object detecting methods, making them more reliable and effective (Zhao et al., 2019). # Part B: Compare and Evaluate AI Techniques

In the problem domain of "Detect and classify birds in video," three AI techniques that can be effectively applied are Convolutional Neural Networks (CNNs), Faster R-CNN, and YOLO (You Only Look Once). Each of these techniques has its strengths and can be tailored to the specific needs of real-time bird detection.

**Convolutional Neural Networks (CNNs)**

Application to Problem Domain: CNNs are suitable for identifying whether a specific object, such as a bird, is present in an image. While CNNs are traditionally used for image classification rather than real-time object detection, they can serve as feature extractors in frameworks like Faster R-CNN or YOLO to detect birds. For example, they identify images with birds in wildlife monitoring systems or environmental research.

However, because CNNs lack boundary boxes and real-time processing capabilities by default, they are less suitable for applications that call for pinpointing the location of birds in a video.

Time to Setup and Produce Results: Setting up CNNs for bird classification is relatively straightforward, especially with transfer learning and pre-trained models such as ResNet or VGG. Given labeled datasets with images of birds, CNNs can produce classification results quickly but lack the speed and spatial awareness required for real-time bird detection in video.

**Faster R-CNN**

Application to Problem Domain: Faster R-CNN excels at detecting birds in complex environments, such as lakes, wetlands, or crowded wildlife scenes. It not only classifies objects as birds but also provides bounding boxes, making it highly effective for applications requiring precise localization. For example, detecting and tracking birds in wildlife videos for population monitoring or behavior analysis.

While Faster R-CNN provides accurate detection, its speed may not be ideal for real-time video applications where rapid frame processing is required, such as in live monitoring systems.

Time to Setup and Produce Results: Faster R-CNN requires significant setup time, including training a custom model on a dataset of annotated bird images with bounding boxes. Despite its longer setup time, it provides high accuracy and detailed detection, which can be critical for applications prioritizing precision over speed.

**YOLO (You Only Look Once)**

Application to Problem Domain: YOLO is ideal for real-time bird detection in video due to its speed and ability to process multiple objects in a single pass. It can effectively localize birds in various environments, such as lakes, parks, or wetlands, while maintaining low latency. For example, detecting birds in drone footage for ecological surveys or live-streaming wildlife cameras.

Its capability to handle real-time detection with high accuracy makes it the best option for applications requiring quick decisions or continuous monitoring.

Time to Setup and Produce Results: YOLO’s setup time is relatively short, especially with pre-trained models. For detecting birds, a custom YOLO model can be trained using a dataset annotated with bird bounding boxes. Once trained, YOLO processes video frames in milliseconds, making it highly efficient for real-time applications.

### Input/Data

**Format of Input Data**
1. Video Input

  Format: .mp4 video file. This is the primary input for object detection. The video should contain moving scenes where birds are visible, and the model will detect and annotate them.

2. Training Data

  Format: The training data consists of images and their corresponding annotations, typically in a text file, that specify the locations of the objects in the images. These images are used to train the model, and the annotations describe the bounding boxes around objects, class labels, and confidence scores. This data is usually stored in a directory structure defined in the configuration file (e.g., config.yaml).

3. Configuration File (YAML)

  Format: .yaml file (YAML format). This file specifies the dataset paths, model parameters, and hyperparameters for training the model. It includes details such as class names, training paths, validation paths, and other training settings.

4. Model Input (Pre-trained YOLO Model)

  Format: .pt (PyTorch model file). This is the pre-trained YOLO model (e.g., yolov10n.pt), which will be fine-tuned during training or used directly for inference (object detection). The model is loaded from a path, located on a cloud storage service (Google Drive).

5. Zip Data File

  Format:.zip file. Contains the dataset used for training the model. This file is transferred and unzipped into the environment using the scp and unzip commands.

**Pre-processing Required**
1. Unzipping the Data

  The .zip file containing the dataset is unzipped to extract the images and annotations. This step ensures that the data is accessible for training and inference.
  Example: !unzip '/content/data.zip' -d '/content/'

2. Video Pre-processing

  The video file is opened and its frames are extracted for processing. Each frame is passed through the YOLO model for object detection.

3. Resizing and Normalization of Images (for training)
  
  During training, input images are often resized to a standard size and normalized to ensure consistency and proper model convergence.

4. Label Encoding (for training)
  
  During training, the class labels (e.g., "bird") in the annotation files are encoded into numerical values. Each class label is mapped to an integer index, which the model will use to learn how to identify that class.

5. Model Training Data Pre-processing (Augmentation)

  Data augmentation techniques (e.g., flipping, rotation, scaling) can be applied to the images to increase the diversity of the training data and improve model robustness.

```python
import cv2
import torch
from ultralytics import YOLOv10 as YOLO

# Load the YOLOv8 model (choose 'yolov8n.pt', 'yolov8s.pt', etc. for different sizes)
model = YOLO('./runs/detect/train/weights/last.pt')  # or another version of YOLOv8 (e.g., yolov8s.pt for small)

# Load the video file
input_video_path = '/content/gdrive/MyDrive/bird_detection/video.mp4'
output_video_path = '/content/gdrive/My Drive/bird_detection/out.mp4'


# Open the video using OpenCV
video_capture = cv2.VideoCapture(input_video_path)

# Get video properties
frame_width = int(video_capture.get(cv2.CAP_PROP_FRAME_WIDTH))
frame_height = int(video_capture.get(cv2.CAP_PROP_FRAME_HEIGHT))
fps = int(video_capture.get(cv2.CAP_PROP_FPS))
total_frames = int(video_capture.get(cv2.CAP_PROP_FRAME_COUNT))

# Define the codec and create VideoWriter object to save output video
fourcc = cv2.VideoWriter_fourcc(*'mp4v')  # Codec
out_video = cv2.VideoWriter(output_video_path, fourcc, fps, (frame_width, frame_height))

# Iterate over each frame
frame_count = 0
while video_capture.isOpened():
    ret, frame = video_capture.read()  # Read a frame
    if not ret:
        break

    # Apply YOLOv8 object detection
    results = model(frame)[0]

    # Iterate through the detections and draw bounding boxes
    for result in results.boxes.data.tolist():  # Each detection in the format [x1, y1, x2, y2, conf, class]
        x1, y1, x2, y2, conf, cls = result[:6]
        label = f'{model.names[cls]} {conf:.2f}'

        # Draw bounding box and label on the frame
        if conf > 0.5:
                cv2.rectangle(frame, (int(x1), int(y1)), (int(x2), int(y2)), (0, 0, 255), 4)  # Bounding box
        # cv2.putText(frame, label, (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, (255, 0, 0), 2)

    # Write the processed frame to the output video
    out_video.write(frame)

    # Print progress
    frame_count += 1
    print(f'Processed frame {frame_count}/{total_frames}')

# Release resources
video_capture.release()
out_video.release()
cv2.destroyAllWindows()

print(f'Output video saved to {output_video_path}')
```

## 1. Appropriate Testing Methods

To evaluate the bird detection model, I applied a combination of quantitative and qualitative testing methods:

  1. Quantitative Testing Using Metrics
  
  Several performance metrics were used to evaluate the effectiveness of the bird detection model:



*   Precision: Measures the proportion of true positive detections (correctly identified birds) out of all detections. A higher precision indicates fewer false positives.
*   Recall: Measures the proportion of true positive detections out of the total number of actual birds in the images. A higher recall indicates fewer missed detections.

*   mAP (mean Average Precision): I evaluated both mAP@0.5 (IoU threshold of 0.5) and mAP@0.5:0.95 (averaged over IoU thresholds from 0.5 to 0.95). These metrics provide a comprehensive measure of how well the model performs across different levels of detection overlap.

2. Dataset Testing

  The model was tested on two types of datasets:



  *   Training Dataset: Used to optimize the model during the training phase.
  *   Validation Dataset: Used to tune hyperparameters and monitor performance after each epoch.

  3. Visual Testing with Output Images and Video
   
   I visually inspected the detection results using both static images and video output. This allowed us to assess the quality of bounding box placement, confidence scores, and overall detection performance under real-world conditions.

  4. Stress Testing on Edge Cases

  The model was tested with challenging scenarios, such as:



*   Birds in cluttered or densely vegetated environments.
*   Birds partially occluded by branches or objects.

*   Low-light or high-glare conditions.
